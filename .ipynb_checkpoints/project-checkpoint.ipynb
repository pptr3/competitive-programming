## Training and Testing 

It is important that, when we choose the traing data, we do not have any anomaly, this because what we want to do is to lean how time serie normally behave in order to indentify non-normal behavior, so we must not learn anomalies.

```python
train_end = pd.to_datetime('2017-06-10 05:00:00')
nab.plot_series(data, labels, test_start=train_end)
```

We then standardize the data, using the mean and the standard deviation from the training set. here we do not what to have any data from the test set because we don't want to have any onfromation about those data. We will use them only at the end of the process.

```python
tmp = data[data.index < train_end] # take only the training data

# standardize
data['value'] = (data['value'] - tmp['value'].mean()) / tmp['value'].std()

data_tr = data[data.index < train_end]
```

```python
data[data.index < train_end]
```

```python
nab.plot_series(data, labels, windows)
```

```python
q1 = data_tr['value'].quantile(0.25)
q3 = data_tr['value'].quantile(0.75)
sigma = data_tr['value'].std()
m =  len(data_tr)
h = 0.9 * min(sigma, (q3-q1) / 1.34) * m**(-0.2)
print(f'The estimated bandwidth is {h}')
```

We can now fit the training data into a KDE estimator. We will use a gaussian kernel.

```python
kde = KernelDensity(kernel='gaussian', bandwidth=h)
kde.fit(data_tr);
```

```python
vmax = data['value'].max()
xr = np.linspace(-3, vmax, 100)
nab.plot_density_estimator_1D(kde, xr)
```

The graph abose is the representation estimated by our KDE estimator, while the one below is the real distribution of our data. We can notice that they are really similar.

```python
vmax = data['value'].max()
nab.plot_histogram(data['value'], vmax=vmax, bins=50)
```

## Alarm signal

```python
# Obtain the log probability density for all the data
ldens = kde.score_samples(data)
# Convert to a pandas series (add the time index)
signal = pd.Series(index=data.index, data=-ldens) # note the "-" sign to negate the logarithm
# Plot the signal over time
nab.plot_series(signal, labels=labels, windows=windows)
```

As we can clearly see we have obtained some really high peaks in proximity of the anomalies. What we want to do now is finding some threshold which will tell us the real anomalies upon all the peaks.

## Choosing the threshold

### Cost model

The costs here below have been discussed with the company because they depend on the company business".

```python
c_alrm = 2 # Cost of investigating a false alarm
c_missed = 10 # Cost of missing an anomaly
c_late =6.5 # Cost for late detection
```

### Effect of the threshold

We can now see how the cost varies when we change the threshold. 

```python
cmodel = nab.ADSimpleCostModel(c_alrm, c_missed, c_late)
thr_range = np.linspace(signal.mean(), signal.max(), 100)
cost_range = [cmodel.cost(signal, labels, windows, thr) for thr in thr_range]
cost_range = pd.Series(index=thr_range, data=cost_range)
nab.plot_series(cost_range)
```

## Validation set

It is important to have a validation set for determining how the results obtained on the traing set will work on unseen data.

```python
val_end = pd.to_datetime('2017-08-23 04:00:00')
nab.plot_series(data, labels, val_start=train_end, test_start=val_end)
```

## Optimize the threshold

```python
signal_opt = signal[signal.index < val_end] # take train + val set
labels_opt = labels[labels < val_end] # extract the position of anomaly
windows_opt = windows[windows['end'] < val_end] # extract thr position of the windows
thr_range = np.linspace(signal.mean(), signal.max(), 100) # range of the thresholds
```

We now test some possible values of the threshold and we want the one which gives the lowest cost.

```python
best_thr, best_cost = nab.opt_thr(signal_opt, labels_opt,
                                  windows_opt,  cmodel, thr_range)
print(f'Best threshold: {best_thr}, corresponding cost: {best_cost}')
```

The cost using the best threshold on the entire dataset is:

```python
ctst = cmodel.cost(signal, labels, windows, best_thr)
print(f'Cost on the whole dataset {ctst}')
```

# Autoencoder

Basic function to create an Autoencoder.

```python
def build_ae(units) :
  ae_x = Input(shape=(1,), dtype='float32')
  ae_z = Dense(units, activation='tanh')(ae_x)
  ae_y = Dense(1, activation='linear')(ae_z)
  ae = Model(ae_x, ae_y)
  ae.compile(metrics=["accuracy"], optimizer='adam', loss='mse')
  return ae

ae = build_ae(24)
```

```python
data_tr = data[data.index < train_end]
data_val = data[(data.index >= train_end) & (data.index < val_end)]
```

```python
es = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, mode='min', restore_best_weights=True)

history = ae.fit(data_tr.values, data_tr.values,
                 validation_data = (data_val.values, data_val.values),
                 callbacks=[es],
                 batch_size=48, epochs=100, verbose=1)
```

```python
preds = pd.DataFrame(index=data.index, columns=data.columns, data=ae(data.values).numpy())
preds.head()
```

```python
se = np.square(data['value'] - preds['value'])
signal_ae = pd.Series(index=data.index, data=se)
nab.plot_series(signal_ae, labels, windows, figsize=nab.figsize)
```

```python
thr_range = np.linspace(signal_ae.quantile(0.5), signal_ae.max(), 100)
signal_opt = signal_ae[signal_ae.index < val_end]
labels_opt = labels[labels < val_end]
windows_opt = windows[windows['end'] < val_end]
best_thr, best_cost = nab.opt_thr(signal_opt, labels_opt,
                                  windows_opt,  cmodel, thr_range)
print(f'Best threshold: {best_thr}, corresponding cost: {best_cost}')
```

```python
ctst = cmodel.cost(signal_ae, labels, windows, best_thr)
print(f'Cost on the whole dataset {ctst}')
```

# RNN

To train out RNN we first need to prepare the data. In order to do so we will use a window with a size of a day so 24 (since we have steps of 1 hour). 

```python
WINDOW_SIZE = 24
features = len(data.columns)

wdata = nab.sliding_window_1D(data, wlen=WINDOW_SIZE) 
```

```python
wdata_out = data.iloc[WINDOW_SIZE:]['value']
wdata_in = wdata.iloc[:-1]
```

```python
wdata_in_tr = wdata_in[wdata_in.index < train_end]
wdata_out_tr = wdata_out[wdata_out.index <= train_end]
```

```python
mask = (wdata_in.index >= train_end) & (wdata_in.index < val_end)
wdata_in_val = wdata_in[mask]
mask = (wdata_out.index > train_end) & (wdata_out.index <= val_end)
wdata_out_val = wdata_out[mask]
```

```python
wdata_in_test = wdata_in[wdata_in.index >= val_end]
wdata_out_test = wdata_out[wdata_out.index > val_end]
```

```python
datasv, labels, windows = nab.load_series(file_name, data_folder, name)
datasv['value'] = (datasv['value'] - datasv['value'].mean()) / datasv['value'].std()
wdata_sv = nab.sliding_window_1D(datasv, wlen=WINDOW_SIZE) 
wdata_out_sv = datasv.iloc[WINDOW_SIZE:]['value']
wdata_in_sv = wdata_sv.iloc[:-1]

x_train = wdata_in_tr.values.reshape(-1, WINDOW_SIZE, features)
x_val = wdata_in_val.values.reshape(-1, WINDOW_SIZE, features)
x_test = wdata_in_test.values.reshape(-1, WINDOW_SIZE, features)

print(f"Train shape: {x_train.shape, wdata_out_tr.shape}")
print(f"Val shape: {x_val.shape, wdata_out_val.shape}")
print(f"Test shape: {x_test.shape, wdata_out_test.shape}")
```

```python
# Build the model
def build_model(units, input_shape, dropout) :
  model = Sequential()
  model.add(Input((input_shape)))
  model.add(LSTM(units, return_sequences=False))
  model.add(Dropout(dropout))
  model.add(Dense(units//2, activation='gelu'))
  model.add(Dense(features, activation='linear'))
  model.compile(optimizer=optimizers.Adam(1e-2), loss='mse')
  return model
```

```python
tf.keras.backend.clear_session()

model = build_model(32, input_shape=x_train.shape[1:], dropout=0.4)
model.summary()
```

```python
# Train the model
es = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, mode='min', restore_best_weights=True)

history = model.fit(x_train, wdata_out_tr, 
                    epochs=100,
                    batch_size=64, 
                    validation_data=(x_val, wdata_out_val),
                    callbacks=[es],
                    shuffle=False)
```

```python
%matplotlib inline

plt.plot(history.history['loss'], label='Training loss')
plt.plot(history.history['val_loss'], label='Validation loss')
plt.legend()
plt.show()
```

## Gaussian Processes

Since we have different part of our data missing with also huge gaps, the methods seen above can't perform well. So we use Gaussian processes which uses some kernel functions in order to define the correlation of the data points and creates a set of observation to obtain conditional distributions.

Since the sklearn library does not handle Date-Time type, we need to convert that time in something else. For semplicity we just create an increasing integer index.

```python
data_dt = ddata.copy()
data_dt['time'] = np.arange(len(data_dt))
```

First we start with a smaller segment so that we can quickly observe how the the kernels behave.

```python
segment = data_dt[(data_dt.index >= '2017-04-01') & (data_dt.index < '2017-05-28')].copy()
nab.plot_series(segment['value'], show_markers=True, figsize=figsize)
```

We create the training set and the validation, paying attention not to shuffle the first and the last points

```python
tmp = segment.dropna()
np.random.seed(42)
idx = np.arange(len(tmp))
np.random.shuffle(idx[1:-1]) # no not shuffle the first/last point
t = idx[1]; idx[1] = idx[-1]; idx[-1] = t # keep first/last points in the left half
sep = 2*len(idx) // 3
trdata = tmp.iloc[idx[:sep]]
tsdata = tmp.iloc[idx[sep:]]
```

Now we build the Gaussian Process Regressor, and we test some kernels in order to observe which combination works better on our data. 

```python
kernel = WhiteKernel(1e-3, (1e-4, 1e-1))
kernel += ConstantKernel(1, (1e-2, 1e2)) * RBF(1, (1e-2, 1e2))
np.random.seed(42)
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10)
gp.fit(trdata[['time']], trdata['value'])
print(gp.kernel_)
```

```python
pred, std = gp.predict(segment[['time']], return_std=True)
pred = pd.Series(index=segment.index, data=pred)
std = pd.Series(index=segment.index, data=std)
```

```python
nab.plot_gp(target_samples=tsdata['value'], samples=trdata['value'], pred=pred, std=std, figsize=figsize)
```

In order to have some king of benchmark we calculate the Log Likelihood od the validation set. And every time that we makethekernel more complicated we compare the log likelyhood, the higher the better.

```python
from scipy.stats import norm
# Obtain predictions for the validation data
pred_ts = pred[tsdata.index]
std_ts = std[tsdata.index]
ldens = norm.logpdf(tsdata['value'], pred_ts, std_ts)
ll = np.sum(ldens)
print(f'Log likelihood of the validation set: {ll:.2f}')
```

We also try with the Matern kernel which is similar to the RBF kernel but allows us to control the smoothnes of the function.

```python
kernel = WhiteKernel(1e-3, (1e-4, 1e-1))
kernel += ConstantKernel(1, (1e-2, 1e2)) * Matern(1, (1e-2, 1e2), 1.5)
np.random.seed(42)
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10)
gp.fit(trdata[['time']], trdata['value'])
print(gp.kernel_)
```

```python
pred, std = gp.predict(segment[['time']], return_std=True)
pred = pd.Series(index=segment.index, data=pred)
std = pd.Series(index=segment.index, data=std)
```

```python
nab.plot_gp(target_samples=tsdata['value'], samples=trdata['value'], pred=pred, std=std, figsize=figsize)
```

```python
# Obtain predictions for the validation data
pred_ts = pred[tsdata.index]
std_ts = std[tsdata.index]
ldens = norm.logpdf(tsdata['value'], pred_ts, std_ts)
ll = np.sum(ldens)
print(f'Log likelihood of the validation set: {ll:.2f}')
```

Until now we have not used any periodicity, so we will use a ExpSineSquared kernel and we add a daily period because as we can see from our data the pattern repeats itself daily.

```python
kernel = WhiteKernel(1e-4, (1e-5, 1e-3))
kernel += ConstantKernel(1, (1e-2, 1e2)) * Matern(1, (1e-2, 1e2), 1.5)
kernel += ConstantKernel(1, (1e-2, 1e2)) * ExpSineSquared(1, 24, (1e-2, 1e2), (20, 28))
#kernel += ConstantKernel(1, (1e-2, 1e2)) * DotProduct(1, (1e-05, 1e5))
np.random.seed(42)
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10)
gp.fit(trdata[['time']], trdata['value'])
print(gp.kernel_)
```

```python
pred, std = gp.predict(segment[['time']], return_std=True)
pred = pd.Series(index=segment.index, data=pred)
std = pd.Series(index=segment.index, data=std)
```

```python
nab.plot_gp(target_samples=tsdata['value'], samples=trdata['value'], pred=pred, std=std, figsize=figsize)
```

```python
# Obtain predictions for the validation data
pred_ts = pred[tsdata.index]
std_ts = std[tsdata.index]
ldens = norm.logpdf(tsdata['value'], pred_ts, std_ts)
ll = np.sum(ldens)
print(f'Log likelihood of the validation set: {ll:.2f}')
```

We can see that we have done a pretty good job with the last kernel as our Log Likelihood is much greater than the previous.

Now we use this kernel on the entire dataset and see the results.

```python
gp2 = GaussianProcessRegressor(kernel=gp.kernel_, optimizer=None)
```

```python
tmp = data_dt.dropna() # The whole series (NaNs excluded)
gp2.fit(tmp[['time']], tmp['value']);
```

```python
pred_all, std_all = gp2.predict(data_dt[['time']], return_std=True)
pred_all = pd.Series(index=data_dt.index, data=pred_all)
std_all = pd.Series(index=data_dt.index, data=std_all)
```
